```python
def preprocess_genres(df, unique_genres):
    temp = df['genres'].apply(lambda x: str(x).lower().split('|'))
    genres_df = list()
    [genres_df.append([1 if xi in t else 0 for xi in unique_genres]) for t in temp]
    genres_df = pd.DataFrame(genres_df, columns=['genre_'+g for g in unique_genres])
    return pd.concat([df, genres_df], axis=1).drop('genres', axis=1)
```

```python
def preprocess_content_rating(df, encoder):
    col = ['content_rating_'+cr for cr in encoder.categories_[0]]
    df_cr = pd.DataFrame(encoder.transform(df['content_rating'].values.reshape(-1,1)), columns = col)
    return pd.concat([df, df_cr], axis=1).drop('content_rating', axis=1)
```

```python
def get_log_col(data_train, verbose=False):
    log_col = list()
    for col in data_train:
        try:
            if 0. in data_train[col].values:
                raise Exception('zero cannot take log')
            log_feats = np.log(data_train[col])
            log_col.append(col)
            if verbose:
                plt.figure(figsize=(20,3))
                plt.subplot(121)
                data_train[col].hist(bins=50, alpha=0.6)
                plt.title(col)
                plt.subplot(122)
                log_feats.hist(bins=50, alpha=0.6)
                plt.title('{} on log scale'.format(col))
                plt.show()
        except:
            if verbose:
                print('{} cannot take log'.format(col))
                plt.clf()
    return log_col
```

```python
from sklearn.preprocessing import OneHotEncoder, StandardScaler, MinMaxScaler

def preprocess_data(train, test, thresh_cutoff=7.2):
    data_train, data_test = train.copy(), test.copy()
    y_train = data_train['imdb_score']
    y_test = data_test['imdb_score']
    data_train = data_train.drop('imdb_score', axis=1)
    data_test = data_test.drop('imdb_score', axis=1)
    
    # log transform
    log_col = get_log_col(data_train)
    data_train[log_col] = np.log(data_train[log_col])
    
    # fillna
#     filled_mode = data_train['content_rating'].mode()[0]
#     filled_median = data_train.median()
    filled_mean = data_train.mean()
#     data_train['content_rating'] = data_train['content_rating'].fillna(filled_mode)
#     data_test['content_rating'] = data_test['content_rating'].fillna(filled_mode)
    nan_col = [col for col, nan_counts in data_train.isna().sum().items() if nan_counts != 0]
    nan_col_test = [col for col, nan_counts in data_test.isna().sum().items() if nan_counts != 0]
    data_train[nan_col] = data_train[nan_col].fillna(filled_mean[nan_col])
    data_test[nan_col_test] = data_test[nan_col_test].fillna(filled_mean[nan_col_test])
    
    # preprocess genre
#     data_train = preprocess_genres(data_train)
#     data_test = preprocess_genres(data_test)
    
    # preprocess onehot content_rating
#     onehot_encoder = OneHotEncoder(handle_unknown='ignore', sparse=False).fit(data_train['content_rating'].values.reshape(-1,1))
#     data_train = preprocess_content_rating(data_train, onehot_encoder)
#     data_test = preprocess_content_rating(data_test, onehot_encoder)

    # standard scaling
    scaling_col = [col for col in data_train.columns if not('genre' in col or 'content' in col)]
#     scaler = StandardScaler().fit(data_train[scaling_col])
#     data_train[scaling_col] = pd.DataFrame(scaler.transform(data_train[scaling_col]), columns=scaling_col)
#     data_test[scaling_col] = pd.DataFrame(scaler.transform(data_test[scaling_col]), columns=scaling_col)
    scaler = MinMaxScaler().fit(data_train)
    data_train = pd.DataFrame(scaler.transform(data_train), columns=data_train.columns)
    data_test = pd.DataFrame(scaler.transform(data_test), columns=data_test.columns)
    
    # preprocess imdb
    y_train = y_train.apply(lambda x: 1 if x >= thresh_cutoff else 0)
    y_test = y_test.apply(lambda x: 1 if x >= thresh_cutoff else 0)

    return data_train, data_test, y_train, y_test
```

```python
log_col = get_log_col(data_train)
df_train, df_test, y_train, y_test = preprocess_data(data_train, data_test)
```

## Logistic Regression

```python
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score, classification_report

def evaluate(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)
    accuracy = accuracy_score(y_true, y_pred)
    cr = classification_report(y_true, y_pred)
    return cr, cm, f1, accuracy

def plot_cm(cm, norm=True):
    if norm:
        cm = cm/cm.sum(axis=0)
    sns.heatmap(cm, annot=True)
    plt.xlabel('Ground Truth')
    plt.ylabel('Prediction')
    plt.title('Confusion Matrix')
    plt.show()
```

```python
cr,  cm, f1, acc = evaluate(y_test, y_pred)
plot_cm(cm)
print(cr)
```
